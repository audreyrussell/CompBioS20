# Effects of Parameter Value Changes on the Lotka-Volterra Model

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 10000
a = 0.6 # bunny reproduction
b = 1.2 #bunny predation 
c = 0.5 #reproductive ratio of foxes per 1 bunny eaten
d = 1 #fox deaths
dt = 0.001 #i think this is just a constant???

possible_as = [0.2, 0.6, 1.0, 1.4]
possible_bs = [1, 1.2, 1.4, 1.6]
possible_cs = [0.2, 0.5, 0.8, 1]
possible_ds = [0.5, 1, 1.5, 2]

x = np.zeros(nsteps)
y = np.zeros(nsteps)
```

```python
# first with variations on a (bunny repdroduction)

fig1 = plt.figure (figsize = (16, 5));
fig2 = plt.figure(figsize = (5,5));
ax2 = plt.axes()

for itrial in range(len(possible_as)):
    a = possible_as[itrial]
    x[0] = 1
    y[0] = 1
    
    for t in range(1,nsteps):
        x[t] = x[t-1] + dt * (a * x[t-1] - b * x[t-1] * y[t-1])
        y[t] = y[t-1] + dt * (c * x[t-1] * y[t-1] - d * y[t-1])
        
    ax1 = fig1.add_subplot(1, len(possible_as), itrial + 1);
    ax1.plot(x, label = "bunnies");
    ax1.plot(y, label = "foxes");
    ax1.legend(loc = "upper right");
    
    ax2.plot(x, y, label = a);
ax2.legend(loc = "upper right");
```

```python
# second: variations on b (bunny predation)

fig1 = plt.figure (figsize = (16, 5));
fig2 = plt.figure(figsize = (5,5));
ax2 = plt.axes()


for itrial in range(len(possible_bs)):
    b = possible_bs[itrial]
    x[0] = 1
    y[0] = 1
    
    for t in range(1,nsteps):
        x[t] = x[t-1] + dt * (a * x[t-1] - b * x[t-1] * y[t-1])
        y[t] = y[t-1] + dt * (c * x[t-1] * y[t-1] - d * y[t-1])
        
    ax1 = fig1.add_subplot(1, len(possible_bs), itrial + 1);
    ax1.plot(x, label = "bunnies");
    ax1.plot(y, label = "foxes");
    ax1.legend(loc = "upper right");
    
    ax2.plot(x, y, label = b);
ax2.legend(loc = "upper right");
```

```python
# third: variations on c (bunny --> fox energy conversion?)

fig1 = plt.figure (figsize = (16, 5));
fig2 = plt.figure(figsize = (5,5));
ax2 = plt.axes()


for itrial in range(len(possible_cs)):
    c = possible_cs[itrial]
    x[0] = 1
    y[0] = 1
    
    for t in range(1,nsteps):
        x[t] = x[t-1] + dt * (a * x[t-1] - b * x[t-1] * y[t-1])
        y[t] = y[t-1] + dt * (c * x[t-1] * y[t-1] - d * y[t-1])
        
    ax1 = fig1.add_subplot(1, len(possible_cs), itrial + 1);
    ax1.plot(x, label = "bunnies");
    ax1.plot(y, label = "foxes");
    ax1.legend(loc = "upper right");
    
    ax2.plot(x, y, label = c);
ax2.legend(loc = "upper right");
```

```python
# last: variations on d (fox death rate)

fig1 = plt.figure (figsize = (16, 5));
fig2 = plt.figure(figsize = (5,5));
ax2 = plt.axes()


for itrial in range(len(possible_ds)):
    d = possible_ds[itrial]
    x[0] = 1
    y[0] = 1
    
    for t in range(1,nsteps):
        x[t] = x[t-1] + dt * (a * x[t-1] - b * x[t-1] * y[t-1])
        y[t] = y[t-1] + dt * (c * x[t-1] * y[t-1] - d * y[t-1])
        
    ax1 = fig1.add_subplot(1, len(possible_ds), itrial + 1);
    ax1.plot(x, label = "bunnies");
    ax1.plot(y, label = "foxes");
    ax1.legend(loc = "upper right");
    
    ax2.plot(x, y, label = d);
ax2.legend(loc = "upper right");
```

Obviously my plan to do a four-part nested loop didn't work out, but I have shamefully accomplished my goal by hand (via lots of copy-and-paste)! Here, we can see the effects the value of each parameter on the equation as it exists with constant values fo the other parameters. 
 An extremely low a (bunny reproduction) value drives the fox population down with little chance of recovery since they have so little to eat. As a increases, though, the fox population begins to fluctuate with that of the bunnies.
 A higher value of b (predation rate coefficient) drives inequality between bunny and fox populationthat is, as it increases, foxes struggle to keep up with the bunny population
 An increase in the c value indicates that bunnies are helping the foxes to reproduce more, as indicated by low fox populations in the very first subplot. Because bunnies are more beneficial to fox populations in the later graphs, they can increase.
 Low d (death rate) in foxes leads to instability in this relationship because there is nothing keeping the bunnies in check. As the death rate increases, so does stability.
