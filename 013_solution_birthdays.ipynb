# solution to the birthday problem

```python
import numpy as np
import matplotlib.pyplot as plt
```

the logic:

* people are showing up one by one (we pretend that they are showing up one by one, as it makes thinking easier)
* once n-th person comes into the room, there are n-a people already there. 
* Let's calculate P(that there's no overlap in birthdays). Each person "claims" a day of the year.
* with n-1 people, it means n-1 days taken. So (n-1)/365 share of the year is taken.
* The p of a new person not falling into the "taken" part of the year is therefore 1-(n-1)/365.
* So we have enough to calculate it. each new person maintains the status quo of no overlap in birthdays with 1-(n-1)/365 probability. And originally, for n = 1, p(no overlap) = 1

```python
maxn=180 # highest number of people we'll consider
p = np.zeros(maxn) #firther answer, n(n). we'll create an array and fill it with zeros.
p[0] = 1 # if no people , p=1
p[1] = 1 # if one person, also no overlap
for i in range(2, maxn):
    p[i] = p[i-1] * (1-(i-1)/365)
#plt.plot(p);
# but the original question was about p(overlap), not p(no overlap)
plt.plot(1-p);
plt.xlabel("# people in the group");
plt.ylabel("P of no shared birthday");
```
