# sensitivity to initial conditions (aka butterfly effect)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sensitivity to two different trajectories
fig = plt.figure()
nsteps = 100
starting_points = [0.6, 0.7]
x = np.zeros(nsteps)
r = 4
# have a loop that goes through different values of r
for m in range(0,5):
    r = m
for j in starting_points:
    x[0] = j
    for i in range(1, nsteps):
        x[i] = r * x[i-1] * (1-x[i-1])
    plt.plot(x);
```

```python
#from class github:

nsteps = 100              # Number of steps in our lil simulation
x = np.zeros(nsteps)      # Number of animals at every step
starting_points = [0.600, 0.60001]

fig = plt.figure(figsize=(14,7))
for x[0] in starting_points:
    r = 3.9                   # Rate of population change (groth - die-out?)

    for i in range(1,nsteps):
        x[i] = r*x[i-1]*(1-x[i-1])

    plt.plot(x);
plt.ylim((0,1)); # Force y axis to be between 0 and 1
```

```python
# prediction horizon
fig = plt.figure()
nsteps = 100
x = np.zeros(nsteps)
r = 2

for i in range(1, nsteps):
    x[i] = (r * x[i-1] * (1-x[i-1]) + 1)
    plt.plot(x);
```

let's say we actually use this equation to predict something.
how far in the future can we reasonably predict?
how to answer this question:

if the question is about time, probably the answer will include a plot that has time as its horizontal axis. 

if we have a plot with time as axis and the question is about prediction quality, then probably the vertical axis should be some kind of prediction quality. 

how to measure rpediction quality?
option 1: use pairs of trajectories, calculate mean(abs(difference)) for each point in time. option 2: use a family of trajectories, noisify/wiggle the starting point, check how much variance it creates at each of the consecutive points

```python
nsteps = 100
ntraces = 10
x = np.zeros(shape=(nsteps,ntraces))
r = 3.9 
wiggle = 0.0001       #the standard deveiation of the original noise in x0

for itrace in range(ntraces): 
    x[0,itrace] = 0.6 + np.random.randn(1) * wiggle
    for istep in range(1,nsteps):
        x[istep, itrace] = r*x[istep-1, itrace] * (1-x[istep-1, itrace])
fig = plt.figure()
plt.plot(np.std(x, axis = 1))
plt.ylim((0,1))
plt.xlabel("time")
plt.ylabel("variability of trajectories");
```
