# growth

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 100 #number of steps in our lil simulation
x = np.zeros(nsteps) #number of animals at every step
x[0] = 0.6 #The original number of animals 
r = 3.45
#rate of population change(growth–die-out?)

for i in range(1,nsteps):
    x[i] = r * x[i-1] * (1-x[i-1])
plt.plot(x, "c");
plt.ylim((0,1)) #forces y axis to be between 0 and 1
plt.xlabel("days?");
plt.ylabel("number of insects");
    
```
